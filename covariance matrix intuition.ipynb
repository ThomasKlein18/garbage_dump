```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# mess around with these values to see what happens
covariance_matrix = [[9.747, 9.371],
                     [9.371, 9.818]]

# creating 100 random points according to a 2d-gaussian as 
# specified by the cov-matrix. The mean has two coordinates.
xmean = 6
ymean = 6
points = np.random.multivariate_normal(mean=(xmean,ymean), 
                        cov=covariance_matrix, size=(100))
# points will have shape (100,2)

plt.figure(figsize=(12,12))

# drawing lines for the means
plt.axhline(y=ymean, c="red")
plt.axvline(x=xmean, c="red")

# setting the limits of the coordinate system
plt.xlim(0,12)
plt.ylim(0,12)

# plotting all points: x-coordinates against y-coordinates
plt.scatter(points[:,0], points[:,1])

plt.show()
```
